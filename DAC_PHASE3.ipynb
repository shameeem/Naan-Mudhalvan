```python
import pandas as pd
import csv
```

```python
input_file = "/content/statsfinal.csv"
df = pd.read_csv(input_file)
print(df)
```

```python
df = df.dropna()  # Remove rows with missing values
print(df)
```

```python
df['Q-P1'] = df['Q-P1'].astype('int')
df['Q-P2'] = df['Q-P2'].astype('int')
df['Q-P3'] = df['Q-P3'].astype('int')
df['Q-P4'] = df['Q-P4'].astype('int')
df['S-P1'] = df['S-P1'].astype('int')
df['S-P2'] = df['S-P2'].astype('int')
df['S-P3'] = df['S-P3'].astype('int')
df['S-P4'] = df['S-P4'].astype('int')
```

```python
df['Date'] = df['Date'].str.strip()
print(df)
```

```python
output_file = "cleaned_file.csv"
df.to_csv(output_file, index=False)
print(output_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned CSV file
input_file = "cleaned_file.csv"
df = pd.read_csv(input_file)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['S-P1'], bins=20, color='blue')
plt.title('Histogram of a Numeric Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['S-P1'], df['S-P2'], color=['red'], alpha=0.5)
plt.title('Scatter Plot of Two Numeric Columns')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
